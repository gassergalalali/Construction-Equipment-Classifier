```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
import sklearn.neural_network
import sklearn.model_selection
import sklearn.metrics
import seaborn as sns
import copy
import pickle

X = np.load("X.npy")
t = np.load("y.npy")

number_of_folds = 5
best_acc = 99999
best_mlp = None
kf = sklearn.model_selection.KFold(number_of_folds, shuffle=True, random_state=1)
for i, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"Fold {i+1} of {number_of_folds}:")
    mlp = sklearn.neural_network.MLPClassifier(
        [1000, 1000, 1000, 1000], 
        early_stopping=True,
        learning_rate_init=0.01,
        tol=0.01,
        verbose=True
    )
    mlp.fit(X[train_index], t[train_index])
    score = mlp.score(X[test_index], t[test_index])
    print(f"The score is {score:.2%}")
    predict = mlp.predict(X[test_index])
    print("The accuracy is:")
    acc = sklearn.metrics.accuracy_score(t[test_index], predict)
    print(acc)
    conf = sklearn.metrics.confusion_matrix(t[test_index], predict)
    print("Confusion Patrix")
    print(conf)
    plt.figure()
    plt.imshow(conf, cmap="Blues", aspect='equal')
    plt.colorbar()
    plt.title(f"Confusion Matrix\nFold{i+1}")
    plt.show()
    if acc < best_acc:
        best_acc = acc
        best_mlp = copy.deepcopy(mlp)
mlp = copy.deepcopy(best_mlp)

file_name = "test_mlp.pickle"
with open(file_name, "wb") as file:
    pickle.dump(mlp, file)
```
